## Ejercicio 3

Muestra un gráfico de líneas con tres líneas de distintos colores y grosores. Cada una de ellas representará los valores de las funciones trigonométricas seno, coseno y tangente entre los 0 u 720 grados. Además, muestra un título y las etiquetas (labels):

```python
grados = np.arange(0.0, 720.0)
seno = np.sin(grados*np.pi/180)
coseno = np.cos(grados*np.pi/180)
tangente = np.tan(grados*np.pi/180)

# Seno
plotsen = sns.lineplot(data=seno, label="Seno", color="blue")
# Coseno
plotcos = sns.lineplot(data=coseno, label = "Coseno", color = "green")
# Tangente
plottan = sns.lineplot(data=tangente, label = "Tangente", color = "red", linewidth=.75)

plottan.set(ylim=[-1.5,1.5])
```

## Ejercicio 7

Muestra en un gráfico de dispersión dos conjuntos de datos aleatorios generados por tí. Estos datos deben ser valores comprendidos entre el 0 y el 25, tanto en el eje X como en el eje Y. Cada conjunto se pintará de un color distinto:

```python
ax = np.random.randint(25, size = [10])
ay = np.random.randint(25, size = [10])
bx = np.random.randint(25, size = [10])
by = np.random.randint(25, size = [10])
cx = np.random.randint(25, size = [10])
cy = np.random.randint(25, size = [10])

dataa = {'x': ax, 'y': ay, 'dt':'a'}
dfa = pd.DataFrame(dataa)
datab = {'x': bx, 'y': by, 'dt':'b'}
dfb = pd.DataFrame(datab)
datac = {'x': cx, 'y': cy, 'dt':'c'}
dfc = pd.DataFrame(datac)
df = dfa.append(dfb).append(dfc)

scat = sns.scatterplot(data=df, x='x', y='y', hue='dt')

scat.set(title="Gráfico de dispersión")
```

# Ejercicio 8

Muestra un gráfico circular estilizado con la distribución de tus horas del día:

```python
# No incluido en Seaborn, se hace con MatPlotLib

datos = [8, 4, 1, 1, 1, 2, 7]
actividades = ["Dormir", "Clase", "Karate", "Comida", "Transporte", "Teatro", "Ocio"]
explosion = [.10,.10,.10,0,0,.10,.10]

# Podemos usar Seaborn para la paleta de colores entre otros
colors = sns.color_palette('dark')

plt.pie(datos, labels = actividades, colors = colors, startangle = 90,
        shadow = True, explode = explosion)

plt.title("Gráfico circular")
plt.show()
```
