## Ejercicio 3

Muestra un gráfico de líneas con tres líneas de distintos colores y grosores. Cada una de ellas representará los valores de las funciones trigonométricas seno, coseno y tangente entre los 0 y los 720 grados. Además, muestra un título y las etiquetas (labels):

```python
# Usar las matematicas para obtener lso valores de senos cosenos tangentes
grados = np.arange(0.0, 720.0)
seno = np.sin(grados*np.pi/180)
coseno = np.cos(grados*np.pi/180)
tangente = np.tan(grados*np.pi/180)

# Seno
plt.plot(seno, label = "Seno", color = "blue")
# Coseno
plt.plot(coseno, label = "Coseno", color = "green")
# Tangente
plt.plot(tangente, label = "Tangente", color = "red")

plt.title("Funciones Trigonométricas")
plt.ylim([-1.5,1.5])
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
```

## Ejercicio 8

Muestra un gráfico circular estilizado con la distribución de tus horas del día:
