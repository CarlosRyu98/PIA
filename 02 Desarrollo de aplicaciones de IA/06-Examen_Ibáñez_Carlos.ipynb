Mostrar del contenido del dataset los siguientes apartados:

*   Una descripción básica de su contenido

## Ejercicio 7

Partiendo del mismo dataset y juego de la primera fila del ejercicio anterior, Muestra los mismos datos que antes en la gráfica de barras (a excepción del campo "Global_Sales") solo que esta vez en un gráfico circular. Como título del gráfico vuelve a utilizar el nombre del videojuego. En cada porción del círculo muestra los porcentajes con el formato "%1.1f%%" (parámetro autopct):

## Ejercicio 8

Partimos del dataset "06-IMDB_Movie_Data.csv" que contiene la siguiente información por columnas:

*   **Rank**: Posición en el ranking
*   **Title**: Título de la película
*   **Genre**: Género o géneros separados por comas
*   **Description**: Descripción corta
*   **Director**: Nombre del director
*   **Actors**: Actores separados por comas
*   **Year**: Año de publicación
*   **Runtime**: Duración (minutos)
*   **Rating**: Media de la nota de los votos de los usuarios
*   **Votes**: Número de votos
*   **Revenue**: Ingresos producidos por la película (millones)
*   **Metascrore**: Un promedio agregado de puntajes críticos. Los valores están entre 0 y 100. Las puntuaciones más altas representan críticas positivas.

Muestra una gráfica que consideres relevante. Para ello, **deja en el código con comentarios todo el trabajo exploratorio que hagas** sobre el dataset:

```python
df8 = pd.read_csv("06-IMDB_Movie_Data.csv")

# Queremos saber el número de películas que se ha sacado cada año

# Para empezar miramos cuantos años comprende la lista para saber si los datos mostrados son visualizables

print("Número de años en el dataset", df8.Year.nunique()) # Número de años en el dataset
print("Años diferentes únicos en el dataset:", df8.Year.unique()) # Años diferentes en el dataset
print("Número de películas cada año:")
print(df8.Year.value_counts(sort=False)) # Número de películas cada año
                                         # Con el sort=False hacemos que value_counts no nos ordene el resultado, así se queda en el mismo orden que unique
# Una vez observamos que el número de años es asequible, decidimos cómo queremos mostrar los datos
# Si quisieramos ver el porcentaje de películas que supone cada año respecto al total, igual nos sería más útil utilizar el gráfico pie
# Decidimos que solo queremos ver el número películas por cada año separado

x = df8.Year.unique() # Valores eje x
y = df8.Year.value_counts(sort=False) # Valores eje y
w = 0.8 # Ancho de la barra

plt.bar(x, y, w)

plt.title("Número de películas por año")
plt.xlabel("Año")
plt.xticks(x) # Nos aseguramos de que el gráfico nos indique todos los años (Si no puede que en el eje solo nos escriba algunos)
plt.ylabel("Número de películas")

plt.show()
```
