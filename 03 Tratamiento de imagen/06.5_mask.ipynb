```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def colorToGrayscale2(img, r=0.2989, g=0.587, b=0.114):
    grayimg = img.copy()
    grayimg = (r*grayimg[:,:,0] + g*grayimg[:,:,1] + b*grayimg[:,:,2])/255 # Funci√≥n para pasar rgb a grayscale
    return grayimg
```

```python
def aplicaMask(matriz, mask):
    matrizMaskada = matriz.copy()
    f = 0
    for fila in matrizMaskada:
        c = 0
        for columna in fila:
            matrizMaskada[f,c] = masking(matrizMaskada, f, c, mask)
            c += 1
        f += 1
    return matrizMaskada

def masking(img, f, c, mask):
    ret = 0
    for x in range(-1, 2, 1):
        for y in range(-1, 2, 1):
            if (f >= img.shape[0]-1): f -= 1
            if (c >= img.shape[0]-1): c -= 1
            ret += img[f+x, c+y] * mask[x+1, y+1]
    return ret

mask = np.array([[1/6, 0, 1/6],[1/6, 0, 1/6],[1/6, 0, 1/6]])

img = colorToGrayscale2(plt.imread("01_panda1.jpg"))
plt.imshow(img, cmap="gray")
plt.show()

img2 = aplicaMask(img.copy(), mask)
plt.imshow(img2, cmap="gray")
plt.show()
```
