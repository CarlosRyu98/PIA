## Ejercicio 9

Investiga cómo transformar una imagen de color a escala de grises sin utilizar ninguna librería. Define una función de obtenga una imagen a color como parametro y devuelva su versión en escala de grises:

```python
def colorToGrayscale(img):

    grayimg = img.copy()

    grayimg = (0.299*grayimg[:,:,0] + 0.587*grayimg[:,:,1] + 0.114*grayimg[:,:,2])/255 # Función para pasar rgb a grayscale

    return grayimg

img9 = plt.imread("portada-panda.jpg")

plt.imshow(colorToGrayscale(img9), cmap="gray")

plt.show()
```
