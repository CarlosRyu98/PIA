No hace falta que únicamente utilicéis los valores 0 y 1 (o 0 y 255), pero os recomiendo como mucho diferenciar 3 o 4 valores distintos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importar imágenes y pasar a escala de grises

```python
def colorToGrayscale2(img, r=0.2989, g=0.587, b=0.114):
    grayimg = img.copy()
    grayimg = (r*grayimg[:,:,0] + g*grayimg[:,:,1] + b*grayimg[:,:,2])/255 # Función para pasar rgb a grayscale
    return grayimg
```

```python
img0 = colorToGrayscale2(plt.imread("01_kamina.jpg"))
plt.imshow(img0, cmap="gray")
plt.show()
img1 = colorToGrayscale2(plt.imread("05_dirk.jpg"))
plt.imshow(img1, cmap="gray")
plt.show()
img2 = colorToGrayscale2(plt.imread("01_panda1.jpg"))
plt.imshow(img2, cmap="gray")
plt.show()
img3 = colorToGrayscale2(plt.imread("01_panda2.jpg"))
plt.imshow(img3, cmap="gray")
plt.show()
img4 = colorToGrayscale2(plt.imread("05_codegeass.jpg"))
plt.imshow(img4, cmap="gray")
plt.show()
img5 = colorToGrayscale2(plt.imread("05_homestuck.webp"))
plt.imshow(img5, cmap="gray")
plt.show()
img6 = colorToGrayscale2(plt.imread("05_kankri.jpg"))
plt.imshow(img6, cmap="gray")
plt.show()
img7 = colorToGrayscale2(plt.imread("05_peblo.jpg"))
plt.imshow(img7, cmap="gray")
plt.show()
img8 = colorToGrayscale2(plt.imread("05_spiderverse.webp"))
plt.imshow(img8, cmap="gray")
plt.show()
img9 = colorToGrayscale2(plt.imread("05_ttgl.jpg"))
plt.imshow(img9, cmap="gray")
plt.show()
```

Función para cambiar valores

```python
def escalado(img):
    copy = img.copy()
    f = 0
    for filas in copy:
        c = 0
        for columnas in filas:
            if columnas < 0.25:
                copy[f,c] = 0
            elif columnas < 0.5 and columnas > 0.25:
                copy[f,c] = 0.4
            elif columnas < 0.75 and columnas > 0.5:
                copy[f,c] = 0.8
            elif columnas > 0.75:
                copy[f,c] = 1
            c += 1
        f += 1
    return copy
```

```python
img10 = escalado(img0)
plt.imshow(img10, cmap="gray")
plt.show()
img11 = escalado(img1)
plt.imshow(img11, cmap="gray")
plt.show()
img12 = escalado(img2)
plt.imshow(img12, cmap="gray")
plt.show()
img13 = escalado(img3)
plt.imshow(img13, cmap="gray")
plt.show()
img14 = escalado(img4)
plt.imshow(img14, cmap="gray")
plt.show()
img15 = escalado(img5)
plt.imshow(img15, cmap="gray")
plt.show()
img16 = escalado(img6)
plt.imshow(img16, cmap="gray")
plt.show()
img17 = escalado(img7)
plt.imshow(img17, cmap="gray")
plt.show()
img18 = escalado(img8)
plt.imshow(img18, cmap="gray")
plt.show()
img19 = escalado(img9)
plt.imshow(img19, cmap="gray")
plt.show()
```
