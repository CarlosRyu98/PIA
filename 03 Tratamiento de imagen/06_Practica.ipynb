Función para pasar a grayscale

```python
def colorToGrayscale2(img, r=0.2989, g=0.587, b=0.114):
    grayimg = img.copy()
    grayimg = (r*grayimg[:,:,0] + g*grayimg[:,:,1] + b*grayimg[:,:,2])/255 # Función para pasar rgb a grayscale
    return grayimg
```

## Ejercicio 2

Haz una función que, recibiendo una imagen en escala de grises (2D) como parámetro, devuelva la imagen tras aplicar una segmentación manual definida por tí (puedes pasárle los parámetros que consideres):

```python
def manual(img, val):
    copy = img.copy()
    f = 0
    for filas in copy:
        c = 0
        for columnas in filas:
            if columnas < val:
                copy[f,c] = 0
            elif columnas >= val:
                copy[f,c] = 1
            c += 1
        f += 1
    return copy
```

Para ello, utiliza el método de la **Binarización de Otsu**.
```python
def otsu(img):
    copy = img.copy()

    ### copiao para ver si entiendo algo
    hist, bin_edges = np.histogram(copy, bins=256)
    bin_mids = (bin_edges[:-1] + bin_edges[1:]) / 2.
    weight1 = np.cumsum(hist)
    weight2 = np.cumsum(hist[::-1])[::-1]
    mean1 = np.cumsum(hist * bin_mids) / weight1
    mean2 = (np.cumsum((hist * bin_mids)[::-1]) / weight2[::-1])[::-1]
    inter_class_variance = weight1[:-1] * weight2[1:] * (mean1[:-1] - mean2[1:]) ** 2
    index_of_max_val = np.argmax(inter_class_variance)
    threshold = bin_mids[:-1][index_of_max_val]
    print(threshold)
    val = threshold
    ###
    
    f = 0
    for filas in copy:
        c = 0
        for columnas in filas:
            if columnas < val:
                copy[f,c] = 0
            elif columnas >= val:
                copy[f,c] = 1
            c += 1
        f += 1
    return copy

img0 = colorToGrayscale2(plt.imread("01_kamina.jpg"))
plt.imshow(img0, cmap="gray")
plt.show()
# plt.hist(img0)
# plt.show()
img30 = otsu(img0)
plt.imshow(img30, cmap="gray")
plt.show()
```
