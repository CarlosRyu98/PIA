```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def colorToGrayscale2(img, r=0.2989, g=0.587, b=0.114):
    grayimg = img.copy()
    grayimg = (r*grayimg[:,:,0] + g*grayimg[:,:,1] + b*grayimg[:,:,2])/255 # Funci√≥n para pasar rgb a grayscale
    return grayimg
```

```python
def aplicaMask(matriz, mask):
    matrizMaskada = matriz.copy()
    f = 0
    for fila in matrizMaskada:
        c = 0
        for columna in fila:
            matrizMaskada[f,c] = s(matrizMaskada, f, c, mask)
            c += 1
        f += 1
    return matrizMaskada

def s(img, f, c, mask):
    ret = 0
    for x in range(-1, 1, 1):
        for y in range(-1, 1, 1):
            ret += img[f+x, c+y] * mask[x+1, y+1]
    return ret


mask = np.array([[1/6, 1/6, 0],[1/6, 1/6, 0],[1/6, 1/6, 0]])
img = plt.imread("01_panda1.jpg")
plt.imshow(img)
plt.show()

img2 = colorToGrayscale2(img)
plt.imshow(img2, cmap="gray")
plt.show()

img3 = aplicaMask(img2.copy(), mask)
plt.imshow(img3, cmap="gray")
plt.show()
```
