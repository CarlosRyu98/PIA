# Ejemplo Red Neuronal funci√≥n XOR

```python
import numpy as np
from tensorflow import keras as kr
```

```python
# cargamos las 4 combinaciones de las compuertas XOR
training_data = np.array([[0,0],[0,1],[1,0],[1,1]])

# y estos son los resultados que se obtienen, en el mismo orden
target_data = np.array([[0],[1],[1],[0]])
```

```python
# Creamos la red neuronal con 2 capas:
model = kr.models.Sequential()
model.add(kr.layers.Dense(16, input_dim=2, activation='relu'))
model.add(kr.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
model.fit(training_data, target_data, epochs=1000)
```

```python
# evaluamos el modelo
scores = model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
print (model.predict(training_data).round())
```

Vemos que las predicciones se corresponden con las salidas deseadas de la funci√≥n XOR.
