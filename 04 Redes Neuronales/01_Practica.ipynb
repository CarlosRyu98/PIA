## Ejercicio 2

Programa una clase "Neurona" que cuente con los siguientes elementos:

*   Un constructor que reciba como parámetro el número de conexiones que tendrá de entrada la Neurona y un número entero del 1 al 4 que utilizaremos para definir la función de activación de la Neurona. 
Con el primer número, el constructor definirá un atributo "weights" o "pesos" de tipo array Numpy del tamaño recibido más uno (para el Bias).  
Todos los pesos deben ser inicializados con valores aleatorios entre 0 y 1 (el del Bias incluído)

*   Una función "weighted_sum" o "suma_ponderada" que reciba como parámetro un array Numpy de la misma forma (shape) que el de los pesos de la Neurona.  
Este array contendrá los valores recibidos como entrada en la neurona (teniendo en cuenta que la primera posición siempre valdrá 1 ya que será el número por el que multipliquemos el peso del Bias). 
La función devolverá la suma ponderada de las entradas multiplicadas cada una por su peso

*   Una función "step_function" o "funcion_activacion" que reciba como parámetro un array Numpy con los parámetros de entrada de la neurona.  
Desde esta función se utilizará la función de la suma ponderada y se aplicará la función de activación que corresponda según el parámetro recibido en el constructor:
    1. Threshold o Escalonada
    2. Sigmoid o Sigmoide
    3. Hyperbolic_Tangent o Tangente_Hiperbolica
    4. ReLU


```python
class Neurona:
    def __init__(self, input1, input2):
        self.weights = np.random.randint(0, 2, (1, input1+1))
        self.act = input2

    def weighted_sum(self, input):
        input = np.insert(input, 0, 1)
        suma = (self.weights*input).sum()
        return suma

    def step_function(self, input):
        suma = self.weighted_sum(input)
        if self.act == 1:
            metodo = "Threshold"
            if suma <= 0: output = 0
            else: output = 1
        elif self.act == 2:
            metodo = "Sigmoid"
            output = 1/(1+math.exp(-suma))
        elif self.act == 3:
            metodo = "Hyperbolic_Tangent"
            output = (math.exp(suma) - math.exp(-suma)) / (math.exp(suma) + math.exp(-suma))
        elif self.act == 4:
            metodo = "ReLU"
            output = max(0, suma)
        print("Resultado: ", output)
        print("Metodo: ", metodo)
```
